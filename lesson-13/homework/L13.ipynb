```python
from datasets import load_dataset
import string

# Load the dataset
ds = load_dataset("roneneldan/TinyStories")

ds_train = ds['train']
ds_train = ds_train.select(range(10000))

# exact characters to remove
chars_to_remove = '¡¦©«±³»ÂÃâœ˜“”€™'

translation_table = str.maketrans('', '', chars_to_remove)

# mapping function
def clean_text_map_fn(example):
    # translate() is highly efficient for deleting characters
    example['text'] = example['text'].translate(translation_table)
    return example

# Applying the function to the dataset
# The .map() method applies the function in parallel
ds_train_cleaned = ds_train.map(clean_text_map_fn)

# Verification
print("--- Cleaning Complete ---")

# Check if any of the characters are present in the *cleaned* dataset
def check_for_char(text, char):
    return char in text

for char in chars_to_remove:
    # any() with a generator expression to quickly check 10,000 strings
    is_present = any(check_for_char(t, char) for t in ds_train_cleaned['text'])
    print(f"Character '{char}': Still Present? {is_present}")

print("\nYour dataset is now cleaned and stored in the 'ds_train_cleaned' variable.")
```

```python
vocab = set()
for i in range(len(ds_train)):
    vocab.update(set(ds_train_cleaned[i]['text']))
vocab = sorted(vocab)
```
