```python
# 1. Инициализация переменных (начальное состояние)
A = 10
B = 3
factorial_A = 1 # Переменная-накопитель для факториала
i = 1             # Счетчик для цикла

# 2. Вычисление факториала числа A с помощью цикла WHILE
# Это демонстрирует пошаговое изменение состояния (значений factorial_A и i)
while i <= A:
    factorial_A = factorial_A * i  # Инструкция присваивания, изменяющая состояние
    i = i + 1                      # Инструкция присваивания, изменяющая состояние
    
# Вывод результата факториала
print(f"Число A: {A}")
print(f"Число B: {B}")
print(f"Факториал числа A ({A}!) = {factorial_A}")

# 3. Проверка кратности с помощью условного оператора
# Это демонстрирует ветвление выполнения программы
if A % B == 0:
    # 4. Вывод сообщения о кратности (Ветвь 1)
    print(f"Результат кратности: Число A ({A}) кратно числу B ({B})")
else:
    # 4. Вывод сообщения о кратности (Ветвь 2)
    print(f"Результат кратности: Число A ({A}) НЕ кратно числу B ({B})")

# Ожидаемый вывод:
# Число A: 10
# Число B: 3
# Факториал числа A (10!) = 3628800
# Результат кратности: Число A (10) НЕ кратно числу B (3)
```
